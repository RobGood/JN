## find max height for given vertical velocity

my second program, reprised in Python

```python
# first define some constants

massEarth = 5.972e24  # kg  
massRoket = 10        # kg

radiusEarth = 20e6 / 3.14159

G = 6.67408e-11 
```

```python
# system state variables

velocity = 11100  # m/s   initial velocity up    escape velocity is about 11186  m/s
height   =   0   # m     up
time     =   0   # s     from start

# time slice size
dt = 10e-3  # s

printTimeStep = 7200  # change this to control the amount of output
lastPrintTime = time - printTimeStep

maxHeight = height

while height >= 0 and time < 1000000:
    distance = radiusEarth + height
    force = ( G * massEarth * massRoket ) / (distance * distance)
    acceleration = force / massRoket
    
    newTime = time + dt
    newHeight = height + velocity * dt
    newVelocity = velocity - acceleration * dt
    
    if time >= lastPrintTime + printTimeStep:
        print("%7.3fs  %6.1fm  %5.1fm/s -- %.0fm %.1fN %.1fm/s/s" % (time, height, velocity, distance, force, acceleration))
        lastPrintTime = time
    
    (time, height, velocity) = (newTime, newHeight, newVelocity)

    if height > maxHeight:
        maxHeight = height
        maxHeightTime = time
        
print("%.3fs  %.1fm  %.1fm/s -- %.0fm %.1fN %.1fm/s/s" % (time, height, velocity, distance, force, acceleration))
print("max height  %.2fm at %.3fs" % (maxHeight, maxHeightTime))
```
