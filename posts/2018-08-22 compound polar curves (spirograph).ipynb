- two gears
  - a circular gear G1, radius r1, T1 teeth on convex side of circumfrence
    - a reference point defines a zero angle
    - the 'plot point' is at radius r0, angle PH
  - a ring gear G2, radius r2, T2 teeth on concave side of circumfrence
    - a reference point defines a zero angle
- r0 < r1 < r2
- G1 meshes inside G2 with the two reference points closest at time 0
- the meshed teeth ensure no slippage and impose a rational proportionality between G1 and G2 rotation angles

what do we know?

- centre of G1 moves in a circle centred on centre of G2
  - radius r2 - r1, angular velocity w
- as G1 revolves and rotates a curve is plotted tracing the motion of the plot point
  - G1 centre: `r(t) = r2-r1, th(t) = w * t`
  - G1 rotation: `phi(t) = th(t) * (1 - T2/T1) + PH`

do some coordinate transformations...

- G1 plot point's coordinates relative to G1 centre
  - `xp(t) = r0 * sin(phi(t)), yp(t) = r0 * cos(phi(t))`
- G1 centre coordinates relative to G2
  - `xc(t) = r(t) * sin(th(t)), yc(t) = r(t) * cos(th(t))`
- add 'em up and plot
  - `x(t) = xp(t) + xc(t), y(t) = yp(t) + yc(t)`



    

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
r0 = 2
r1 = 35
r2 = 39
T1 = 50
T2 = T1 * r2/r1
PH = 0
w = 2*np.pi / 1000
maxt = 50000
````

````python
def r(t):
    return r2 - r1

def th(t):
    return w * t

def phi(t):
    return th(t) * (1 - T2/T1) + PH
````

````python
def xp(t):
    return  r0 * np.sin(phi(t))

def yp(t):
    return  r0 * np.cos(phi(t))

def xc(t):
    return r(t) * np.sin(th(t))

def yc(t):
    return r(t) * np.cos(th(t))

def x(t):
    return xp(t) + xc(t)

def y(t):
    return yp(t) + yc(t)
````

````python
r0 = 2
r1 = 35
r2 = 39
T1 = 50
T2 = T1 * r2/r1
PH = 0
w = 2*np.pi / 1000
maxt = 50000
````

````python
(X,Y) = ([], [])

for t in range(0,maxt):
    X.append(x(t))
    Y.append(y(t))
    
plt.plot(X, Y)
plt.show()
````
